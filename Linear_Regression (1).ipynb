Linear models are very powerful in their predictive power assuming datasets contain some degree of linear or polynomial relationship between the input features and the values. The intuitive of linear models often arises from the ability to view data as plotted on a graph and observe a tredning pattern in the data with the output (the y axis value for the data) trending positively or negatively with the input (the x axis value). While often not presetend as such, the fundamental components of linear regression models are often learned during high school math classes. It is the equation of straight line or linear model: y = mx+b

Here x is the input value and y is the corresponding output or predicted value. The patameters of the model are the gradient or slop of the line (the change in y values divided by change in x) define by m as well as the y-intercept value b, which indicates where the line crosses the y axis. With such model, we can provide values for the m and b parameters to construct a linear model. For example, y=2x+1 has a slope of 2 indicating the changes in y values are at a rate of twice that of x; the line crosses the y intercept at 1.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

The next step is we need to generate random data with linear relationship. To make it more realistic, we also add a random noise element. 

```python
#Generate a random data
np.random.seed(0)
area = 2.5 * np.random.randn(100) + 25
price = 25 * area + 5 + np.random.randint(20,50, size = len(area))
data = np.array([area, price])
data = pd.DataFrame(data = data.T, columns=['area','price'])
plt.scatter(data['area'], data['price'])
plt.show()
```

Now we will calculate the 2 regression coefficients using the equations which we defined above. 

```python
W = sum(price*(area-np.mean(area))) / sum((area-np.mean(area))**2)
b = np.mean(price) - W*np.mean(area)
print("The regression coefficients are", W,b)
```

Regression coefficients are estimates of the unknown population parameters and describe the relationship between a predictor variable and the response. In linear regression, coefficients are the values that multiply the predictor values 
(source: https://statisticsbyjim.com/glossary/regression-coefficient/)


```python
# We need to try to predice new prices using the weight and bias values
y_pred = W *area+b
```

```python
# Plotting predicted prices along with the actual price
plt.plot(area, y_pred, color='red',label="Predicted Price")
plt.scatter(data['area'], data['price'], label="Training Data")
plt.xlabel("Area")
plt.ylabel("Price")
plt.legend()
```

As you can see from the above image, the predicted values follow the same trend as the actual house prices.
